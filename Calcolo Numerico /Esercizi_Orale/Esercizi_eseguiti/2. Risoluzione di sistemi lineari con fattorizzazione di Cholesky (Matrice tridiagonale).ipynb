### Dato un problema test di dimensioni variabili Ax = b la cui soluzione esatta è il vettore x ̄ = (1, . . . , 1)T e b è il termine noto, ottenuto come Ax ̄ calcolare:
### • il numero di condizionamento della matrice A
### • la soluzione x del sistema lineare Ax = b mediante la fattorizzazione di Cholesky
### Considerare come matrice A la matrice tridiagonale simmetrica definita positiva, avente sulla diagonale principale elementi uguali a 9 ed elementi nella sopra e sottodiagonale uguali a -4, con n variabile.

```python
import numpy as np
import scipy.linalg

n = 4
A = np.eye(n)*9 - np.eye(n, k = -1)*4 - np.eye(n, k = 1)*4
x = np.ones((n, 1))
b = A@x

print("\nA.shape: ", A.shape)
print("x.shape: ",x.shape)
print("b.shape: ", b.shape)
```

```python
CondA = np.linalg.cond(A, 2)
print("\nK(A)= ", CondA)
```

```python
L = scipy.linalg.cholesky(A, lower = True)

print("\nL = ", L)
print("\nL*L.T = ", np.matmul(L, np.transpose(L)))
print('\nErrore fattorizzazione Cholesky = ', scipy.linalg.norm(A-np.matmul(L, np.transpose(L)), 'fro'))
```

```python
y = scipy.linalg.solve_triangular(L, b, lower = True)
my_x = scipy.linalg.solve(np.transpose(L), y, lower = False)

print('\nmy_x = \n ', my_x)
print("\nNorma dell'errore =", np.linalg.norm(x-my_x, 'fro'))
```
