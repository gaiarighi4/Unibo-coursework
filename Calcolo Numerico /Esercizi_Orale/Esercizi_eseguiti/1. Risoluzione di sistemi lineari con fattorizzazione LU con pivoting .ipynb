### Dato un problema test di dimensioni variabili Ax = b la cui soluzione esatta è il vettore x ̄ = (1, . . . , 1)T e b è il termine noto, ottenuto come Ax ̄ calcolare: 
### • il numero di condizionamento della matrice A
### • la soluzione x del sistema lineare Ax = b mediante la fattorizzazione LU con pivoting.
### Considerare come matrice A la matrice quadrata, di numeri casuali generati mediante la funzione randn di Numpy, con n variabile tra 10 e 1000.

```python
import numpy as np
import random 
import scipy.linalg

n = random.randint(10, 1000)
A = np.random.randn(n, n)
x = np.ones((n, 1))
b = A@x

print("\nA.shape: ", A.shape)
print("x.shape: ", x.shape)
print("b.shape: ", b.shape)
```

```python
CondA = np.linalg.cond(A, 2)
print("\nK(A) = ", CondA)
```

```python
lu, piv = scipy.linalg.lu_factor(A)
print('\nlu = ', lu,)
print('\npiv =', piv,)
```

```python
my_x = scipy.linalg.lu_solve((lu, piv), b) 

print("\nmy_x = ", my_x)
print("\nNorma dell'errore = ", np.linalg.norm(x-my_x, 'fro'))
```
